```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df['Age'], kde=True, bins=20)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.countplot(x='ChestPainType', data=df)
plt.title('Chest Pain Type Count')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Sex', y='Cholesterol', data=df)
plt.title('Cholesterol Levels by Sex')
plt.show()
```

```python
sns.scatterplot(x='Age', y='MaxHR', hue='Sex', data=df)
plt.title('Max Heart Rate by Age (Colored by Sex)')
plt.show()
```

```python
numeric_df = df.select_dtypes(include='number')

plt.figure(figsize=(10, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Numeric Features')
plt.show()

```

Summary of EDA Findings
- Age is mostly between 40–60 years.
- Most common chest pain types are ATA and ASY.
- Cholesterol levels vary widely among both males and females.
- Max heart rate tends to decrease as age increases.
- Strong correlation between Age and MaxHR.
